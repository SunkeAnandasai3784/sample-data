```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset (replace with your actual dataset)
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'C', 'A'],
    'Value1': [10, 20, 15, 25, 30, 10, 40, 35, 45, 15],
    'Value2': [5, 15, 10, 20, 25, 5, 30, 25, 35, 10]
}
df = pd.DataFrame(data)

# Function for Relational Plot
def plot_relational(df):
    sns.relplot(x='Value1', y='Value2', hue='Category', data=df)
    plt.title("Relational Plot (Scatter Plot)")
    plt.show()

# Function for Categorical Graph
def plot_categorical(df):
    sns.catplot(x='Category', y='Value1', kind='bar', data=df)
    plt.title("Categorical Plot (Bar Plot)")
    plt.show()

# Function for Statistical Graph
def plot_statistical(df):
    plt.figure(figsize=(8, 5))
    sns.histplot(df['Value1'], kde=True, bins=5)
    plt.title("Statistical Plot (Histogram & KDE)")
    plt.show()

# Function to Display Statistics
def display_statistics(df):
    mean_val = df['Value1'].mean()
    median_val = df['Value1'].median()
    std_dev = df['Value1'].std()
    correlation = df[['Value1', 'Value2']].corr()
    description = df['Value1'].describe()
    skewness = df['Value1'].skew()
    kurtosis = df['Value1'].kurtosis()

    print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Std Dev: {std_dev:.2f}")
    print("\nCorrelation Matrix:\n", correlation)
    print("\nData Description:\n", description)
    print(f"\nSkewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f}")

# Execute all functions
plot_relational(df)
plot_categorical(df)
plot_statistical(df)
display_statistics(df)
```
